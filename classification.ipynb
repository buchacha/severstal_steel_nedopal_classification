```python
# SVM

import pandas as pd
from sklearn.model_selection import train_test_split, ShuffleSplit

data = pd.read_csv('extracted_data/data_hist_30_30_255_rgb.csv', header = None)

print(type(data))
data = data.drop([0])
print(data)
X = data[range(1, data.shape[1])]
y = data[0]
```

```python
from sklearn.svm import SVC

rs = ShuffleSplit(n_splits=10, test_size=0.3)

for train_index, test_index in rs.split(X):
    X_train = X[X.index.isin(train_index)]
    y_train = y[y.index.isin(train_index)]
    X_test = X[X.index.isin(test_index)]
    y_test = y[y.index.isin(test_index)]

    clf = SVC(kernel='poly', C = 1000, random_state=241)
    clf.fit(X_train, y_train)
    print(clf.score(X_test, y_test))
#     print(clf.support_)
```

```python
from sklearn.ensemble import GradientBoostingClassifier

rs = ShuffleSplit(n_splits=10, test_size=0.3)

for train_index, test_index in rs.split(X):
    X_train = X[X.index.isin(train_index)]
    y_train = y[y.index.isin(train_index)]
    X_test = X[X.index.isin(test_index)]
    y_test = y[y.index.isin(test_index)]

    clf = GradientBoostingClassifier(n_estimators=1000, verbose=False)
    clf.fit(X_train, y_train)
    print(clf.score(X_test, y_test))
#     print(clf.support_)
```

```python
from sklearn.ensemble import RandomForestClassifier

rs = ShuffleSplit(n_splits=15, test_size=0.3)

for train_index, test_index in rs.split(X):
    X_train = X[X.index.isin(train_index)]
    y_train = y[y.index.isin(train_index)]
    X_test = X[X.index.isin(test_index)]
    y_test = y[y.index.isin(test_index)]

    clf = RandomForestClassifier(n_estimators=10, verbose=False)
    clf.fit(X_train, y_train)
    print(clf.score(X_test, y_test))
#     print(clf.support_)
```

```python
import numpy as np
from sklearn.model_selection import KFold, cross_val_score

# clf = RandomForestClassifier(n_estimators=1, verbose=False, random_state=42)
# clf = GradientBoostingClassifier(n_estimators=1, verbose=False, random_state=42)
clf = SVC(kernel='poly', C = 1000)

kf = KFold(n_splits=15, shuffle=True)

score = cross_val_score(clf, X, y, cv=kf, scoring='precision')

print(np.mean(score))
```
